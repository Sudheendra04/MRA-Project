```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv("dataset_group.csv")  # Replace with your actual file path
```

```python
# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])
```

```python
# Extract useful time-related features
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Quarter'] = data['Date'].dt.quarter
data['DayOfWeek'] = data['Date'].dt.dayofweek
data['Day'] = data['Date'].dt.day
```

```python
# Grouping and aggregation
monthly_sales = data.groupby(['Year', 'Month'])['Product'].count()
quarterly_sales = data.groupby(['Year', 'Quarter'])['Product'].count()
day_of_week_sales = data.groupby('DayOfWeek')['Product'].count()
daily_sales = data.groupby('Day')['Product'].count()
```

```python
# Plots
plt.figure(figsize=(12, 6))
```

```python
# Monthly Sales
plt.subplot(2, 2, 1)
sns.lineplot(data=monthly_sales.reset_index(), x='Month', y='Product', hue='Year')
plt.title('Monthly Sales Over Years')
```

```python
# Quarterly Sales
plt.subplot(2, 2, 2)
sns.lineplot(data=quarterly_sales.reset_index(), x='Quarter', y='Product', hue='Year')
plt.title('Quarterly Sales Over Years')
```

```python
# Day of Week Sales
plt.subplot(2, 2, 3)
sns.barplot(x=day_of_week_sales.index, y=day_of_week_sales.values)
plt.title('Day of Week Sales')
```
